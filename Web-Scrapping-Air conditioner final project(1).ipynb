# <p style="padding:10px;background-color:#686868;margin:0;color:#c6b9be;font-family:newtimeroman;font-size:150%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500"> Pie Chart</p>

```python
# Pie chart of Reviews distribution
# Counting the occurrences of each unique review
review_counts = {}
for review in reviews:
    if review in review_counts:
        review_counts[review] += 1
    else:
        review_counts[review] = 1

# Extracting labels and values for the pie chart
labels = [f'{review} stars' for review in review_counts.keys()]
values = list(review_counts.values())

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Reviews')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Load data from Excel file
df = pd.DataFrame(data)


# Calculate mean and median
mean_value = df['Reviews'].mean()
median_value = df['Reviews'].median()

# Normal Distribution Plot with Mean and Median Lines
plt.figure(figsize=(10, 6))
sns.distplot(df['Reviews'], fit=norm, kde=False, hist_kws=dict(edgecolor="k", linewidth=2))
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
plt.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_value:.2f}')
plt.title('Normal Distribution Plot with Mean and Median')
plt.legend()
plt.show()
```
